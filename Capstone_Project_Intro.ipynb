# Data

To solve the problem, we need the data consisting of

# Extracting the data


1.	Fortunately, the 78 Denver-neighborhoods data is readily available on the statistical_neighborhoods.csv file in the Denver Mile High City website. The url of the website is https://www.denvergov.org/opendata/dataset/city-and-county-of-denver-statistical-neighborhoods. 
2.	The web scrapping techniques is used to extract the average house rental price in those neighborhoods from the website https://www.rentcafe.com/average-rent-market-trends/us/co/denver/
3.	We use the Geocoder package for obtaining the Latitudes and Longitudes of these neighborhoods.
4.	Finally, we get the venue data related to these neighborhoods using the Foursquare API.
